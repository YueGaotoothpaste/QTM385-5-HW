````python
# Your answers here

class Matrix:

    # Constructor
    def __init__(self, num, nrow): 
        if len(num) % nrow == 0:
            self.nrow = nrow
            self.num = num
            ncol = int(len(num)/nrow)
            self.ncol = ncol
            self.dim = [nrow,ncol]
            self.mat = []
            for i in range(0,len(num),ncol):
                self.mat.append(num[i:i+ncol])
        else:
            raise ValueError('Error. Invalid input.')
        
        pass
    
    # Print represetation
    def __str__(self):
        
        return '\n'.join(['\n'.join(str(self.mat[i]) for i in range(0,self.nrow))])

        pass
    
    # Representation
    def __repr__(self):
        rpinfo = """
        Matrix dimension [{row},{col}]
        """.format(row = self.nrow, col = self.ncol)
        return rpinfo
        
        pass
    
    # Test to see if square matrix
    def is_square(self):
        if len(self.mat)==len(self.mat[0]):
            return True
        else:
            return False
       
        pass
        
    # Addition (see https://en.wikipedia.org/wiki/Matrix_addition)
    def __add__(self, other):
        numadd = []
        if self.nrow == other.nrow and self.ncol == other.ncol:
            for i in range(0,len(self.num)):
                numadd.append(self.num[i]+other.num[i])
            nrowsadd = self.nrow
            output = Matrix(numadd,nrowsadd)
            return output
        else:
            raise ValueError('Error. The matrices do not conform.')
    
        pass
    
    # Subtraction
    def __sub__(self, other):
        numsub = []
        if self.nrow == other.nrow and self.ncol == other.ncol:
            for i in range(0,len(self.num)):
                numsub.append(self.num[i]-other.num[i])
            nrowsub = self.nrow
            output = Matrix(numsub,nrowsub)
            return output
        else:
            raise ValueError('Error. The matrices do not conform.')
        pass
        
    # Scalar multiplication (https://en.wikipedia.org/wiki/Scalar_multiplication)
    def __mul__(self, num):
        nummul = [i * num for i in self.num]
        nrowmul = self.nrow
        output = Matrix(nummul, nrowmul)
        return output
        
        pass
    
    # Extract row
    def exrow(self, i):
        if i < self.nrow:
            return self.mat[i]
        else:
            raise ValueError('Error. Row index out of bounds.')
        
        pass
    
    # Extract column
    def excol(self, i):
        if i < self.ncol:
            return list(self.mat[j][i] for j in range(0,self.nrow))
        else:
            raise ValueError('Error. Column index out of bounds.')
        
        pass
    
    # Matrix multiplication (see https://en.wikipedia.org/wiki/Matrix_multiplication)
    def __pow__(self, other):
        if self.ncol == other.nrow:
            numpow= []
            for i in range(0,self.nrow):
                for j in range(0,other.ncol):
                    sums = 0
                    for k in range(0, self.ncol):
                        sums = sums + self.mat[i][k]*other.mat[k][j]
                    numpow.append(sums)
            nrowspow = self.nrow
            output = Matrix(numpow,nrowspow)
            return output
        else:
            raise ValueError('Error. Invalid matrix input.') 
   
        pass
    
    # Extract submatrix for minor (see https://en.wikipedia.org/wiki/Minor_(linear_algebra))
    def subm(self, row, col):
        if row < self.nrow and col < self.ncol:
            newnrow = self.nrow-1
            lst = []
            newnum = []
            for r in self.mat[:row]+self.mat[row+1:]:   
                lst.append(r[:col] + r[col+1:])
            for i in lst:
                for j in i:
                    newnum.append(j)
            output = Matrix(newnum, newnrow)
            return output
        else:
            raise ValueError('Error. Index out of scope.')
        
        pass
    
    # Determinant (see https://en.wikipedia.org/wiki/Determinant)
    def det(self):
        if self.is_square():
            if self.nrow == 2:
                return self.mat[0][0]*self.mat[1][1]-self.mat[0][1]*self.mat[1][0]
            
            determinant = 0
            for i in range(self.nrow):
                determinant += ((-1)**i)*self.mat[0][i]*(self.subm(0,i)).det()
            return determinant
        else:
            raise ValueError('Error. Not a square matrix.')  
                
        pass
    
    # Transpose
    def t(self):
        newnum = []
        for i in self.mat:
            for j in i:
                newnum.append(0)
        tmat = Matrix(newnum, self.ncol)
        for i in range(self.nrow):
            for j in range(self.ncol):
                tmat.mat[j][i] = self.mat[i][j]
                t = []
        for i in tmat.mat:
            for j in i:
                t.append(j)
        output = Matrix(t, self.ncol)
        return output
       
        pass
    
    # Cofactor matrix (see https://en.wikipedia.org/wiki/Minor_(linear_algebra))
    def cofm(self):
        if self.is_square():
            co = []
            for i in range(self.nrow):
                for j in range(self.ncol):
                    co.append((-1)**(i+j)*(self.subm(i,j)).det())
            output = Matrix(co,self.nrow)
            return output
        else:
            raise ValueError('Error. Not a square matrix.')  
                
        pass
    
    # Inverse matrix (see the inverse definition after buiding the cofactor matrix: 
    #                 https://en.wikipedia.org/wiki/Minor_(linear_algebra))
    def inv(self):
        if self.is_square() and self.ncol == 1:
            raise ValueError('Error. Singular matrix.')
        else:
            output = self.cofm().t()*(1/self.det())
            return output
       
        pass
    
````
